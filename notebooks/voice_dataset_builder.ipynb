```python
import sys
import os

# Make project root visible
sys.path.append(os.path.abspath(".."))

import pandas as pd
from src.feature_extraction.audio_features import extract_features

# Correct dataset path
DATASET_PATH = os.path.abspath("../data/Coswara-Data")

print("Dataset path:", DATASET_PATH)
print("Exists:", os.path.exists(DATASET_PATH))
```

```python
def get_counting_files():
    counting_files = []

    for root, dirs, files in os.walk(DATASET_PATH):
        for file in files:
            if file.endswith(".wav") and "counting" in file:
                full_path = os.path.join(root, file)
                counting_files.append(full_path)

    return counting_files


files = get_counting_files()

print(f"Total counting files: {len(files)}")

files[:5]
```

```python
dataset = []

for i, file in enumerate(files):
    try:
        features = extract_features(file)

        row = {
            "file": file,
            **features
        }

        dataset.append(row)

        if i % 100 == 0:
            print(f"Processed {i}/{len(files)} files")

    except Exception as e:
        continue
```

```python
len(dataset)
```

```python
df = pd.DataFrame(dataset)

df.head()
```

```python
os.makedirs("../data/voice_features", exist_ok=True)

df.to_csv("../data/voice_features/voice_dataset.csv", index=False)

print("Dataset saved!")
```

```python
metadata = pd.read_csv("../data/Coswara-Data/combined_data.csv")

print(metadata.shape)
metadata.head()
```

```python
def extract_user_id(file_path):
    return file_path.split("/")[-2]


# test
df["user_id"] = df["file"].apply(extract_user_id)

df[["file", "user_id"]].head()
```

```python
# Merge dataset with metadata on user_id
df = df.merge(metadata, left_on="user_id", right_on="id", how="left")

# Check result
df[["user_id", "covid_status"]].head()
```

```python
df["covid_status"].value_counts()
```

```python
# Keep only required columns

columns_to_keep = ["file", "user_id", "covid_status"] + [
    col for col in df.columns if col.startswith("mfcc_") or col in [
        "pitch", "spectral_centroid", "zcr", "jitter", "shimmer", "hnr"
    ]
]

df_clean = df[columns_to_keep]

df_clean.head()
```

```python
df_clean.to_csv("../data/voice_features/voice_dataset_labeled.csv", index=False)

print("Final labeled dataset saved!")
```

```python
def map_label(status):
    if status == "healthy":
        return 0
    else:
        return 1


df_clean["label"] = df_clean["covid_status"].apply(map_label)

df_clean["label"].value_counts()
```

```python
df_clean.to_csv("../data/voice_features/voice_dataset_labeled.csv", index=False)

print("Final labeled dataset saved!")
```
